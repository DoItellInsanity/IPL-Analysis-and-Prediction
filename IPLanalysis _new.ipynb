```python
#Making new features
imp_columns_2nd_inn['current_score']=imp_columns_2nd_inn.groupby('match_id')['total_runs_y'].cumsum()
imp_columns_2nd_inn
```

```python
imp_columns_2nd_inn['runs_left'] = imp_columns_2nd_inn['total_runs_x'] - imp_columns_2nd_inn['current_score'] + 1
imp_columns_2nd_inn
```

```python
imp_columns_2nd_inn['balls_left'] = 120 -(imp_columns_2nd_inn['over']*6 + imp_columns_2nd_inn['ball'])
imp_columns_2nd_inn
```

```python
imp_columns_2nd_inn['player_got_out'] = imp_columns_2nd_inn['player_dismissed'].fillna(0,inplace = True)
imp_columns_2nd_inn['player_got_out'] = imp_columns_2nd_inn['player_dismissed'].apply(lambda x: 1 if isinstance(x,str) else 0)
imp_columns_2nd_inn['player_got_out'] = imp_columns_2nd_inn['player_got_out'].astype(int)

#imp_columns_2nd_inn['player_got_out'].astype(float)
wicket = imp_columns_2nd_inn.groupby('match_id')['player_got_out'].cumsum()
imp_columns_2nd_inn['wickets_left'] = 10 - wicket
imp_columns_2nd_inn
```

```python
# Getting Current run rate
imp_columns_2nd_inn['CRR'] = imp_columns_2nd_inn['current_score']*6/ (120 - imp_columns_2nd_inn['balls_left'])
```

```python
# Getting Required Run rate
imp_columns_2nd_inn['RRR'] = (imp_columns_2nd_inn['runs_left'])*6/ imp_columns_2nd_inn['balls_left']
imp_columns_2nd_inn
```

```python
# Getting the winner team
def winning(row):
  return(1 if row['batting_team']==row['winner'] else 0)

# toss winning team
def toss_winning(row):
  return(1 if row['toss_winner']==row['batting_team'] else 0)
```

```python
imp_columns_2nd_inn['bool_winning'] = imp_columns_2nd_inn.apply(winning,axis=1)
imp_columns_2nd_inn['bool_toss_winning'] = imp_columns_2nd_inn.apply(toss_winning,axis=1)
imp_columns_2nd_inn
```
